```python
var('u')
p=point([0,0,0],viewer='tachyon')
for i in range(100,0,-1):
    u=i/5
    p=p+point([u*sin(u), u*cos(u),u ],viewer='tachyon')
    p.save('pics/y-%s.png' %(100-i))

p.show()
p=p+parametric_plot3d((u*sin(u), u*cos(u),u ), (u,1,20),viewer='tachyon')
p.show()
```
